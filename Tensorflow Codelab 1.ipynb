```python
# Importing all of the neccessary resources

import tensorflow as tf
import numpy as np
from tensorflow import keras
```

```python
# Creating a simple newual network with 1 layer and 1 neuron
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

```python
# Creating a model using stochastic gradient descent as the optimizer and mean squared error for loss
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
# Creating the data that will be used through numpy arrays
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-2.0, 1.0, 4.0, 7.0, 10.0, 13.0], dtype=float)
```

```python
# Training the neural network
model.fit(xs, ys, epochs=500)
```

```python
# Forcing the model to make a prediction
print(model.predict([10.0]))
```
